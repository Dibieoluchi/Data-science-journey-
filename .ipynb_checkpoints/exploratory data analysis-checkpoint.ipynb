1. Data Collection:Data collection is the process of gathering and measuring information on variables of interest in a systematic and organized manner. It involves the collection of data points or observations to support research, analysis, or decision-making.


2. Descriptive Statistics: Descriptive statistics involves the use of numerical and graphical methods to summarize and describe the main features of a dataset. This includes measures such as mean, median, mode, range, variance, and standard deviation. Descriptive statistics help in simplifying large amounts of data and making it more understandable.


3. Exploratory Data Analysis (EDA): Exploratory Data Analysis is an approach to analyzing data sets to summarize their main characteristics, often with the help of statistical graphics and other data visualization methods. The primary goal is to gain insights, identify patterns, and generate hypotheses that can inform further analysis or research.


4. Estimation: Estimation in statistics refers to the process of using sample data to make inferences about the characteristics of a population. It involves estimating population parameters, such as means or proportions, based on sample statistics. Common methods of estimation include point estimation and interval estimation.


5. Hypothesis Testing: Hypothesis testing is a statistical method used to make inferences about a population based on a sample of data. It involves formulating a hypothesis about a population parameter, collecting and analyzing data, and then using statistical tests to assess the evidence against or in favor of the hypothesis. The goal is to make decisions about whether to accept or reject a null hypothesis in favor of an alternative hypothesis.

# Histogram

accorrding to this book,  a useful way to describe a variable in a dataset is by showcasing the values it contains and how frequently each value appears. This presentation of values and their frequencies is known as the distribution of the variable. The most common way to represent this distribution is through a histogram, a graph illustrating the frequency of each value.

There are multiple ways to compute frequencies in Python

# Using a Dictionary:

This is a basic and manual way to count frequencies, Manually iterate through the dataset and use a dictionary to store the frequency of each unique value.

for x in t:

    hist[x] = hist.get(x, 0) + 1

```python
live_births # now  i have a new dataframe containing data when preganacy outcome is of live birth 
```

```python
# Next I generate and plot the histogram of birthwgt_lb , that is birth weight in pounds  for live births for my case i will be using the value counts method 
```

```python
import matplotlib.pyplot as plt 
```

```python
# Use value_counts to get the frequency of each unique value in 'birthwgt_lb' column 
birth_weight_counts = live_births['birthwgt_lb'].value_counts(sort=True)
```

```python
# Plot the histogram
birth_weight_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel(' Birth Weight (lb)')
plt.ylabel('Frequency')
plt.title('Histogram of Birth Weight for Live Births')
plt.show()
```

```python
live_births['birthwgt_lb'].plot(kind='kde', color='skyblue', linewidth=5)
plt.xlabel('Total Birth Weight (lb)')
plt.ylabel('Density')
plt.title('Kernel Density Estimate of  Birth Weight for Live Births')
plt.show()
```

The histogram of birth weights shows that the most frequent value is 7 pounds, making it the mode of the distribution of live birthweights . The kernel density plot shows that the overall shape is roughly bell-shaped, resembling a normal distribution also known as gaussian Distribution . However, unlike a perfect bell curve, this distribution is asymmetrical, with a longer tail on the left side than on the right.
